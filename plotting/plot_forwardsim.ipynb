```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib.backends.backend_pdf import PdfPages
p1 = PdfPages('forward_sim.pdf')
import math

### Visualizing forward simulation 

time_lim = 15
states_true = np.load('../run_77/saved_forwardsim/forwardsim_states_true_0.npy')
states_pred = np.load('../run_77/saved_forwardsim/forwardsim_states_pred_0.npy')
print (states_pred.shape)
print (states_true.shape)



plt.figure()
plt.title('x pos')
plt.plot(states_true[0:time_lim, 0])
plt.plot(states_pred[0:time_lim, 0], 'r--', linewidth=2)
plt.savefig(p1, format='pdf')

plt.figure()
plt.title('y pos')
plt.plot(states_true[0:time_lim, 1])
plt.plot(states_pred[0:time_lim, 1], 'r--', linewidth=2)
plt.savefig(p1, format='pdf')

plt.figure()
plt.title('z pos')
plt.plot(states_true[0:time_lim, 2])
plt.plot(states_pred[0:time_lim, 2], 'r--', linewidth=2)
plt.savefig(p1, format='pdf')

plt.figure()
plt.title('x vel')
plt.plot(states_true[0:time_lim, 3])
plt.plot(states_pred[0:time_lim, 3], 'r--', linewidth=1.5)
plt.savefig(p1, format='pdf')

plt.figure()
plt.title('y vel')
plt.plot(states_true[0:time_lim, 4])
plt.plot(states_pred[0:time_lim, 4], 'r--', linewidth=1.5)
plt.savefig(p1, format='pdf')

plt.figure()
plt.title('z vel')
plt.plot(states_true[0:time_lim, 5])
plt.plot(states_pred[0:time_lim, 5], 'r--', linewidth=1.5)
plt.savefig(p1, format='pdf')

plt.figure()
plt.title('roll')
roll_true = np.arctan2(states_true[0:time_lim, 7], states_true[0:time_lim, 6])
roll_pred = np.arctan2(states_pred[0:time_lim, 7], states_pred[0:time_lim, 6])
plt.plot(roll_true[0:time_lim])
plt.plot(roll_pred[0:time_lim], 'r--')
plt.savefig(p1, format='pdf')

plt.figure()
plt.title('pitch')
roll_true = np.arctan2(states_true[0:time_lim, 9], states_true[0:time_lim, 8])
roll_pred = np.arctan2(states_pred[0:time_lim, 9], states_pred[0:time_lim, 8])
plt.plot(roll_true[0:time_lim])
plt.plot(roll_pred[0:time_lim], 'r--')
plt.savefig(p1, format='pdf')

plt.figure()
plt.title('yaw')
roll_true = np.arctan2(states_true[0:time_lim, 11], states_true[0:time_lim, 10])
roll_pred = np.arctan2(states_pred[0:time_lim, 11], states_pred[0:time_lim, 10])
plt.plot(roll_true[0:time_lim])
plt.plot(roll_pred[0:time_lim], 'r--')
plt.savefig(p1, format='pdf')

plt.figure()
plt.title('wx')
plt.plot(states_true[0:time_lim, 12])
plt.plot(states_pred[0:time_lim, 12], 'r--')
plt.savefig(p1, format='pdf')

plt.figure()
plt.title('wy')
plt.plot(states_true[0:time_lim, 13])
plt.plot(states_pred[0:time_lim, 13], 'r--')
plt.savefig(p1, format='pdf')

plt.figure()
plt.title('wz')
plt.plot(states_true[0:time_lim, 14])
plt.plot(states_pred[0:time_lim, 14], 'r--')
plt.savefig(p1, format='pdf')

plt.figure()
plt.title('left')
left_true = np.arctan2(states_true[0:time_lim, 16], states_true[0:time_lim, 15])
left_pred = np.arctan2(states_pred[0:time_lim, 16], states_pred[0:time_lim, 15])
left_true[left_true<0]=2*math.pi+left_true[left_true<0] ###make it from 0 to 2pi instead of -pi to pi
left_pred[left_pred<0]=2*math.pi+left_pred[left_pred<0] ###make it from 0 to 2pi instead of -pi to pi
plt.plot(left_true[0:time_lim])
plt.plot(left_pred[0:time_lim], 'r--')
plt.savefig(p1, format='pdf')

plt.figure()
plt.title('right')
right_true = np.arctan2(states_true[0:time_lim, 18], states_true[0:time_lim, 17])
right_pred = np.arctan2(states_pred[0:time_lim, 18], states_pred[0:time_lim, 17])
right_true[right_true<0]=2*math.pi+right_true[right_true<0] ###make it from 0 to 2pi instead of -pi to pi
right_pred[right_pred<0]=2*math.pi+right_pred[right_pred<0] ###make it from 0 to 2pi instead of -pi to pi
plt.plot(right_true[0:time_lim])
plt.plot(right_pred[0:time_lim], 'r--')
plt.savefig(p1, format='pdf')

plt.figure()
plt.title('left vel')
plt.plot(states_true[0:time_lim, 19])
plt.plot(states_pred[0:time_lim, 19], 'r--')
plt.savefig(p1, format='pdf')

plt.figure()
plt.title('right vel')
plt.plot(states_true[0:time_lim, 20])
plt.plot(states_pred[0:time_lim, 20], 'r--')
plt.savefig(p1, format='pdf')

plt.figure()
plt.title('bemf r')
plt.plot(states_true[0:time_lim, 21])
plt.plot(states_pred[0:time_lim, 21], 'r--')
plt.savefig(p1, format='pdf')

plt.figure()
plt.title('bemf l')
plt.plot(states_true[0:time_lim, 22])
plt.plot(states_pred[0:time_lim, 22], 'r--')
plt.savefig(p1, format='pdf')

plt.figure()
plt.title('vbat')
plt.plot(states_true[0:time_lim, 23])
plt.plot(states_pred[0:time_lim, 23], 'r--')
plt.savefig(p1, format='pdf')

p1.close()
```
